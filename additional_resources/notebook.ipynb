```python
import matplotlib.pyplot as plt

# Inicializando a lista dos valores parciais da série
partial_sums = []
current_sum = 0

N = 10

# Calculando os primeiros 9 termos da série
for i in range(1, N + 1):
    current_sum += 1 / (2**i)
    partial_sums.append(current_sum)

# Plotando apenas os pontos da série
plt.plot(range(1, N + 1), partial_sums, marker='o', linestyle='')

# Anotação dos valores nos pontos
for i, value in enumerate(partial_sums):
    plt.text(i + 1, value + 0.05, f"{value:.3f}", ha='center', va='bottom')

plt.title('Série: $\\sum_{n=1}^{\\infty} \\frac{1}{2^n}$')
plt.xlabel('Termo (n)')
plt.ylabel('$S_n$ : Valor da Soma Parcial')

# Ajustando os ticks do eixo x para exibir de 1 em 1
plt.xticks(range(1, N + 1))

# Habilitando o grid
plt.grid(True)

# Ajustando os limites dos eixos
plt.ylim(0.35, 1.25)
plt.xlim(0, 11)

# Exibindo o gráfico

plt.savefig('serie_1')
```

```python
import matplotlib.pyplot as plt
import math

# Inicializando a lista dos valores parciais da série
partial_sums = []
current_sum = 0

N = 10

# Calculando os primeiros 9 termos da série
for i in range(1, N + 1):
    current_sum += math.log(i / (i + 1))
    partial_sums.append(current_sum)

# Plotando apenas os pontos da série
plt.plot(range(1, N + 1), partial_sums, marker='o', linestyle='')

# Anotação dos valores nos pontos
for i, value in enumerate(partial_sums):
    plt.text(i + 1, value + 0.05, f"{value:.3f}", ha='center', va='bottom')

plt.title('Série: $\\sum_{n=1}^{\\infty} log (\\frac{n}{n + 1})$') 
plt.xlabel('Termo (n)')
plt.ylabel('$S_n$ : Valor da Soma Parcial')

# Ajustando os ticks do eixo x para exibir de 1 em 1
plt.xticks(range(1, N + 1))

# Habilitando o grid
plt.grid(True)

# Ajustando os limites dos eixos
plt.ylim(-3, 0)
plt.xlim(0, 11)

# Exibindo o gráfico

plt.savefig('serie_2')
```

```python
import matplotlib.pyplot as plt
 

# Inicializando a lista dos valores parciais da série
partial_sums = []
current_sum = 0

N = 25

# Calculando os primeiros 9 termos da série
for i in range(1, N + 1):
    current_sum += (3 / (i * (i + 1))) + (1 / 2**i)   
    partial_sums.append(current_sum)

# Plotando apenas os pontos da série
plt.plot(range(1, N + 1), partial_sums, marker='o', linestyle='')

# Anotação dos valores nos pontos
for i, value in enumerate(partial_sums):
    if i in (0, 5, 10, 15, 20, 24):
        plt.text(i + 1, value + 0.05, f"{value:.3f}", ha='center', va='bottom')

plt.title('Série: $\\sum_{n=1}^{\\infty} \\frac{3}{n(n + 1)} + \\frac{1}{2^n} $') 
plt.xlabel('Termo (n)')
plt.ylabel('$S_n$ : Valor da Soma Parcial')

# Ajustando os ticks do eixo x para exibir de 1 em 1
plt.xticks(range(1, N + 1))

# Habilitando o grid
plt.grid(True)

# Ajustando os limites dos eixos
plt.ylim(1.5, 4.5)
plt.xlim(-1, 27)

# Exibindo o gráfico

plt.savefig('serie_3')
```
